```python
# Definindo uma seed padrão para os exemplos

rng = np.random.default_rng(seed= 1701) # Seed para reprodução.
```

```python
# Definindo arrays de exemplo

x1 = rng.integers(10, size= 6) # Array unidimensional
x2 = rng.integers(10, size= (3, 4)) # Array de duas dimensões, linhas e colunas.
x3 = rng.integers(10, size= (3, 4, 5)) # Array tidimensional
```

```python
# Alguns atributos que cada um dos arrays possui

print("x3 ndim: ", x3.ndim)
print("x3 shape: ", x3.shape)
print("x3 size: ", x3.size)
print("x3 dtype: ", x3.dtype)
```

```python
# Sub arrays unidimensionais.

x1
```

```python
# Elementos depois do índice três

x1[3:]
```

```python
# Acessando cada segundo array, partindo do primeiro item

x1[1::2]
```

```python
# Quando o step é negativo o start e o end são trocados.

x1[::-1] # Agora começamos no último elementos até o primeiro
```

```python
# Saltando de dois em dois do fim para começo

x1[::-2]
```
