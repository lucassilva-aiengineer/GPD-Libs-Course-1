```python
# Criando um array

# Arrays são muito interessantes para evitarmos os gastos que são implementados
# pela linguagem python em listas.
```

```python
# Criando Arrays

lista_inteiros = list(range(1000))
```

```python
print(lista_inteiros)
```

```python
array_1 = np.array(lista_inteiros.append("uma string"))
```

```python
print(array_1.dtype)
```

```python
# Ditando o tipo de dados do array na criação

array_3 = np.array([1, 2, 3, 4, 5.2], dtype= np.int32)
```

```python
array_3 # O valor após o ponto é simplesmente ignorado
```

```python
array_4 = np.array([1, 2, 3, 4.5, 4.10, 1.30], dtype= np.float16)
```

```python
array_4
```

```python
# Criando um array numpy multidimensional, partindo de listas aninhadas.

matriz = np.array([range(i, i + 3) for i in [2, 4, 6]], dtype= np.int8)
```

```python
matriz.dtype
```

```python
# Criando Arrays com as rotinas predefinidas do numpy

vetor_zeros = np.zeros(15, dtype= int)
```

```python
vetor_zeros
```

```python
vector_ones = np.ones(10, dtype= np.int8)
```

```python
vector_ones
```

```python
# Matriz ones

# 3 x 5 = Três linhas e cinco colunas
matriz_ones = np.ones((3, 5), dtype= np.int8)
```

```python
matriz_ones
```

```python
# np.full

# Preeche uma matriz específicada com um valor específicado

matriz_pi = np.full((3, 5), 3.14)
```

```python
matriz_pi
```

```python
# Criando sequências numéricas, algo que lembra uma progressão aritmética simples.
# Imprime o intervalo de numeros entre o 0 e o 10 saltando de 2 em dois, muito parecido com o range nativo python.
pares = np.arange(0, 10, 2)
```

```python
print(pares)
```

```python
# Criado uma sequência de valores igualmente espaçados entre 0 e 1.


np.linspace(0, 1, 5)
```

```python
# Criando uma Matriz 3x3 de valores pseudoaleatórios
# entre 0 e 1
np.random.random((3, 3))
```

```python
# Criando uma matriz de valores de média 0 e desvio de padrão 1

matriz_media_zero = np.random.normal(0, 1, (3, 3))
```
